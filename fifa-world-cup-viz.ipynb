visualize countries who won the cup

```python
plt.figure(figsize=(10,8))
sns.countplot(x='Winner',data=worldcups)
plt.show()
```

Wow! Brazil has won about 5 times in world cup history!

```python
plt.figure(figsize=(10,8))
plt.hist(worldcups['Country'])
plt.xticks(rotation=90)
#plt.ylim([0,1])
plt.show()
```

3 countries hosted the world cup 3 times!

```python

plt.figure(figsize=(15,10))
sns.catplot(x='Country',y='GoalsScored',hue='Year', data=worldcups,kind='bar',height=8)
plt.xticks(rotation=90)
#plt.ylim([0,1])
plt.show()
```
